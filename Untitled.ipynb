```python
# Columns to exclude
removable_columns = ["india_adani", "nepal", "wind", "solar"]

df["other"] = df[removable_columns].fillna(0).sum(axis=1)

df.drop(columns=removable_columns + ["remarks"], inplace=True)
```

```python
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    return df[(df[column] >= lower) & (df[column] <= upper)]

cols = ["liquid_fuel"]

for col in cols:
    df = remove_outliers(df, col)
```

```python
# Mathematically sound anomaly detection using IQR
Q1 = df['demand_mw'].quantile(0.25)
Q3 = df['demand_mw'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Capping outliers to the bounds instead of removing to maintain time-series continuity
df['demand_mw'] = df['demand_mw'].clip(lower=lower_bound, upper=upper_bound)
```

```python
weather_data["time"] = pd.to_datetime(weather_data["time"])
weather_data = weather_data.sort_values("time")

weather_30min = (
    weather_data
    .set_index("time")
    .resample("30min")
    .interpolate(method="time")
    .reset_index()
)

df["datetime"] = pd.to_datetime(df["datetime"])

df = df.merge(
    weather_30min,
    left_on="datetime",
    right_on="time",
    how="left"
)

df = df.sort_values('datetime')

df["year"] = df["datetime"].dt.year
df["month"] = df["datetime"].dt.month
df["day"] = df["datetime"].dt.day
df["hour"] = df["datetime"].dt.hour
df["minute"] = df["datetime"].dt.minute
df["day_of_week"] = df["datetime"].dt.dayofweek
df["is_weekend"] = df["day_of_week"].isin([5,6]).astype(int)
df['lag_1'] = df['demand_mw'].shift(1)
df['lag_2'] = df['demand_mw'].shift(2)
df['lag_24'] = df['demand_mw'].shift(24)
df['rolling_mean_3'] = df['demand_mw'].shift(1).rolling(3).mean()
df['rolling_mean_24'] = df['demand_mw'].shift(1).rolling(24).mean()
df['rolling_std_24'] = df['demand_mw'].shift(1).rolling(24).std()


df.drop(columns=["datetime", "time", "generation_mw"], inplace=True, errors="ignore")

print(df.head())
```

```python
source_cols = [
    "gas", "liquid_fuel", "coal", "hydro",
     "india_bheramara_hvdc",
    "india_tripura","source_sum", "other"
]

df=df.drop(columns=source_cols)
print(df.head())

df['target_next_hour'] = df['demand_mw'].shift(-1)
df.dropna(inplace=True)

X_test = df[df["year"]==2023].drop(columns=["target_next_hour","load_shedding"])
Y_test =df[df["year"]==2023]["target_next_hour"]

X_train = df[df["year"]!=2023].drop(columns=["target_next_hour","load_shedding"])
Y_train =df[df["year"]!=2023]["target_next_hour"]

# Verification
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape} (All rows should be from 2023)")
```

```python


from sklearn.preprocessing import StandardScaler, PowerTransformer

# Transform X (Yeo-Johnson)
pt = PowerTransformer(method="yeo-johnson")
X_train= pt.fit_transform(X_train)
X_test=pt.transform(X_test)

```

```python

# Scale X
scaler = StandardScaler()
X_train= scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_percentage_error
import numpy as np

model = LinearRegression()
model.fit(X_train, Y_train)
y_pred = model.predict(X_test)

r2 = r2_score(Y_test, y_pred)
rmse = np.sqrt(mean_squared_error(Y_test, y_pred))
mape = mean_absolute_percentage_error(Y_test, y_pred)

print("Linear Regression")
print("Average R2:", r2)
print("Average RMSE:", rmse)
print("mape", mape)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=100, n_jobs=-1)
model.fit(X_train, Y_train)
y_pred = model.predict(X_test)

r2 = r2_score(Y_test, y_pred)
rmse = np.sqrt(mean_squared_error(Y_test, y_pred))
mape = mean_absolute_percentage_error(Y_test, y_pred)

print("Randome forest regression ")
print("Average R2:", r2)
print("Average RMSE:", rmse)
print("mape", mape)
```
